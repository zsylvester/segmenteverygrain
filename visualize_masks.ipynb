```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.colors import ListedColormap
from PIL import Image

%matplotlib inline

CMAP = ListedColormap(['black', 'steelblue', 'orange'])
LEGEND_PATCHES = [
    mpatches.Patch(color='black',     label='0 – background'),
    mpatches.Patch(color='steelblue', label='1 – grain interior'),
    mpatches.Patch(color='orange',    label='2 – grain boundary'),
]
```

```python
# Set the paths to the masks you want to visualize
mask_paths = [
    './prediction_outputs/cropped_prac7_etched_096_sam_mask.png',
    './prediction_outputs/cropped_prac7_etched_096_unet_mask.png',
]
```

```python
for mask_path in mask_paths:
    mask = np.array(Image.open(mask_path).convert('L'))

    fig, ax = plt.subplots(figsize=(10, 8))
    ax.imshow(mask, cmap=CMAP, vmin=0, vmax=2, interpolation='nearest')
    ax.set_title(mask_path.split('/')[-1])
    ax.axis('off')
    ax.legend(handles=LEGEND_PATCHES, loc='lower right', framealpha=0.8, fontsize=9)
    plt.tight_layout()
    plt.show()
```
