```python
#thanks again chat
import os

# === User settings ===
input_dir = "cropped"   # directory containing images
output_file = "file_list_cropped_test.txt"        # name of the output text file

# === Script ===
with open(output_file, "w") as f:
    for filename in sorted(os.listdir(input_dir)):
        # Skip hidden/system files
        if filename.startswith('.'):
            continue
        # Only include regular files (not directories)
        full_path = os.path.join(input_dir, filename)
        if os.path.isfile(full_path):
            f.write(filename + "\n")

print(f"Saved {output_file} with {sum(1 for _ in open(output_file))} filenames.")
```
